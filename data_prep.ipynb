```python
movies['cast'] = credits['cast']
movies['crew'] = credits['crew']
movies.head()
```

```python
def get_actor(x, n):
    x = ast.literal_eval(x)
    if len(x) > n - 1:
        return x[n - 1].get('name')
    else:
        return np.nan

movies['actor_1_name'] = movies['cast'].apply(get_actor, n = 1)
movies['actor_2_name'] = movies['cast'].apply(get_actor, n = 2)
movies['actor_3_name'] = movies['cast'].apply(get_actor, n = 3)
movies.head()
```

```python
def get_director(x):
    x = ast.literal_eval(x)
    for i in x:
        if i.get('job') == 'Director':
            return i.get('name')

movies['director_name'] = movies['crew'].apply(get_director)
movies.head()
```

```python
def get_genre_list(x):
    x = ast.literal_eval(x)
    genres = ''
    for i in x:
        if i.get('name') == 'Science Fiction':
            genres += 'Sci-Fi, '
        else:
            genres += i.get('name') + ', '
    return genres[:-2]

movies['genres'] = movies['genres'].apply(get_genre_list)
movies.head()
```

```python
def create_data(year, num_of_table):
    # Fetch the HTML content
    url = f'https://en.wikipedia.org/wiki/List_of_American_films_of_{year}'
    response = requests.get(url)
    from tmdbv3api import Movie
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.select('table')
    table
    df = pd.read_html(str(table), header = 0)
    data = df[2]
    for i in range(3,num_of_table + 1):
        data = pd.concat([data,df[i]])
    data.reset_index(drop = True)
    data.drop(['Opening', 'Opening.1','Ref.'], axis = 1, inplace = True)
    data.reset_index(drop = True)
    tmdb_movie = Movie()
    def get_actor_and_director(x):
        try:
            if x == np.nan:
                return [np.nan] * 4
            x = x.split('; ')
            director = x[0]
            director = director[:director.find('(') - 1]
            actors = x[-1].split(', ')[:3]
            actors += [np.nan] * (3 - len(actors))
            actor_1, actor_2, actor_3 = actors
            return [director, actor_1, actor_2, actor_3]
        except:
            data['Cast and crew'][data['Cast and crew'] == x] = np.nan
    
    def get_genre(x): 
        genres = []
        try:
            result = tmdb_movie.search(x) 
            movie_id = result[0].id 
            response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key)) 
            data_json = response.json()
            if data_json['genres']:
                genre_str = " " 
                for i in range(0,len(data_json['genres'])):
                    genres.append(data_json['genres'][i]['name'])
                return genre_str.join(genres)
            else:
                np.nan
        except:
            data['Title'][data['Title'] == x] = np.nan
    data['genres'] = data['Title'].apply(get_genre)
    data[['director_name','actor_1_name','actor_2_name','actor_3_name']] = data['Cast and crew'].apply(get_actor_and_director).apply(pd.Series)
    data = data.drop(['Production company', 'Cast and crew'], axis = 1)
    data['comb'] = data['actor_1_name'] + ', ' + data['actor_2_name'] + ', '+ data['actor_3_name'] + ', '+ data['director_name'] + ', ' + data['genres']
    data.rename(columns = {'Title':'movie_title'}, inplace = True)
    data = data[new_data.columns]
    return data.reset_index(drop = True)

```
